```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import urllib
from bs4 import BeautifulSoup
import re
import time
import pandas as pd
```

```python
salary_df = pd.DataFrame(columns = ['YEAR', 'RK','NAME','TEAM','SALARY'])
salary_url = 'http://www.espn.com/nba/salaries/_/year/{yr}/page/{pgno}/seasontype/4'
# data available from 2000-2020
year_list = range(2000, 2021)
```

```python
for year in year_list:
    curr_pg = 1
    html = urllib.request.urlopen(salary_url.format(yr=str(year), pgno=curr_pg))
    soup = BeautifulSoup(html, 'lxml')
    
    # get the last page for the year
    pg_cnt = soup.find_all('div', {'class':'page-numbers'})[0].getText()
    pg_cnt = pg_cnt.split(' ')
    last_pg = int(pg_cnt[2])
    
    for curr_pg in range(1, last_pg+1):
        html = urllib.request.urlopen(salary_url.format(yr=str(year), pgno=curr_pg))
        print(salary_url.format(yr=str(year), pgno=curr_pg))
        soup = BeautifulSoup(html, 'lxml')
        salary_tables = soup.find_all('table')[0]
        rows = salary_tables.find_all('tr')
        all_rows = []
        for row in rows:
            tds = row.find_all('td')
            store = [td.getText() for td in tds]
            if store != ['RK','NAME','TEAM','SALARY']:
                tds = [year] + store
                all_rows.append(tds)
            else:
                pass
            
        for i in range(len(all_rows)):
            idx = len(salary_df)
            salary_df.loc[idx] = all_rows[i]
        
        time.sleep(2)

print(salary_df)
```

```python
salary_df.to_csv('player_salaries.csv', index=False)
```
